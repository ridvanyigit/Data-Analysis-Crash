# Free Full HD Desktop Wallpaper Download from Unsplash API

##Importing Libraries: `requests` & `urllib.request`

```python
import requests
import urllib.request as url
```

Fetching Data from Unsplash API

```python
r = requests.get('https://api.unsplash.com/search/collections?query=forest&page=18per_page=30&client_id=m2nsNlSmZz74GdCFNYNHy1YddJ29X5kWhT8w8RYssoU')
```

Parsing JSON Response

Examining Keys in the JSON Data

##Downloading Images from 'results'

```python
for img_data in data['results']:
    file_name = str(img_data['id'])+'.jpg'
    img_url = img_data['cover_photo']['urls']['raw']
    suffix = '&q=80&fm=jpg&crop=entropy&cs=tinysrgb&w=1920&fit=max'
    print(img_url)
    img_url = img_url + suffix
    url.urlretrieve(img_url, file_name)
```
