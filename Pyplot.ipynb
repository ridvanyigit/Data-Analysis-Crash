```python
df = pd.DataFrame(np.random.randn(100, 3), columns=['A', 'B', 'C'])
print(df.head())
df['A'] = df['A'].cumsum() + 10
df['B'] = df['B'].cumsum() + 20
df['C'] = df['C'].cumsum() + 30
print(df.head())
```

```python
df.shape
```

```python
df.iplot()
```

```python
df.iplot(x='A', y='B', mode='markers', title='Random Data')
```

```python
plt.plot(df)
plt.show()
```

```python
df.iplot(x='A', y='B', mode='markers', size=25)
```
