```python
plt.subplot(1, 2, 1) # Creates a one-row, two-column grid of plots and selects the first plot.
plt.plot(x, y, 'r--') # Plots x against y with dashed red lines.

plt.subplot(1, 2, 2) # Selects the second subplot in the same 1x2 grid.
plt.plot(y, x, 'g*-') # Plots y against x with green star-markers.
```

```python
plt.subplot(2, 1, 1) # Creates a two-row, one-column grid of plots and selects the first plot.
plt.plot(x, y, 'b--') # Plots x against y with blue color.

plt.subplot(2, 1, 2) # Selects the second subplot in the same 2x1 grid.
plt.plot(y, x, 'y*-', markersize=20) # Plots y against x with yellow color.
```

> figure size and labels

```python
fig, ax = plt.subplots(figsize = (8, 5), dpi = 100)

ax.plot(x, np.sin(y), color = 'r', label = 'sin(y)')
ax.plot(x, np.sqrt(y), color = 'b', label = 'sqrt(y)')

ax.set_xlabel('X Axis', color = 'b')
ax.set_ylabel('Y Axis', color = 'g')
ax.set_title('Set Title of Random', color = 'y')

ax.legend(loc= 0)
```

```python
fig, ax = plt.subplots(1, 2, figsize = (8, 5))

ax[0].plot(x, y, x, y**2)

ax[1].plot(x, np.exp(x))
ax[1].set_yscale('log')

fig.tight_layout()
```

```python
fig, ax = plt.subplots(figsize = (8, 5))

ax.plot(x, y**2)
ax.set_xlabel('X Axis', color = 'r')
ax.set_xticks([1,2,3,4,5])
ax.set_xticklabels([r'$\alpha$', r'$\beta$', r'$\gamma$', r'$\delta$', r'$\epsilon$'], fontsize = 20)

ax.set_ylabel('Y Axis', color = 'g')
ax.set_yticks([0, 100, 200, 300, 400, 500, 600, 700, 800, 900, 1000])
ax.set_yticklabels(['I', 'II', 'III', 'IV', 'V', 'VI', 'VII', 'VIII', 'IX', 'X', 'XI'])

plt.show()
```
